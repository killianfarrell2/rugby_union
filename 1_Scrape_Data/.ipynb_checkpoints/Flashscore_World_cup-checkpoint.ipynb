```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.edge.service import Service
from selenium.webdriver.edge.options import Options
import time

# Set up Edge options
edge_options = Options()
#edge_options.add_argument("--headless")  # Run in headless mode
edge_options.add_argument("--disable-gpu")
edge_options.add_argument("--no-sandbox")
edge_options.add_argument("--disable-dev-shm-usage")

# Set up the Edge WebDriver
webdriver_path = 'C:\\Users\\killi\\OneDrive\\Desktop\\edgedriver_win64\\msedgedriver.exe'
service = Service(webdriver_path)

# Create driver again - this opens up web page
driver = webdriver.Edge(service=service, options=edge_options)

# Navigate to a simple website (e.g., example.com)
url = 'https://www.flashscore.com/rugby-union/world/world-cup/results/'
driver.get(url)

# Allow the page to load
time.sleep(2)

# List to store filtered links
match_summary_links = []


# Initialize lists to store the extracted data
home_team_ls = []
away_team_ls = []
home_score_ls = []
away_score_ls = []
date_ls = []

# Locate matches
matches = driver.find_elements(By.CSS_SELECTOR, 'div.event__match')

# Loop through each match and extract data
for match in matches:
    # Extract home and away team names
    home_team = match.find_element(By.CSS_SELECTOR, 'div.event__participant.event__participant--home').text
    home_team_ls.append(home_team)
    
    away_team = match.find_element(By.CSS_SELECTOR, 'div.event__participant.event__participant--away').text
    away_team_ls.append(away_team)
    
    # Extract home and away scores
    home_score = match.find_element(By.CSS_SELECTOR, 'div.event__score.event__score--home').text
    home_score_ls.append(home_score)
    
    away_score = match.find_element(By.CSS_SELECTOR, 'div.event__score.event__score--away').text
    away_score_ls.append(away_score)

    date = match.find_element(By.CSS_SELECTOR, 'div.event__time').text
    date_ls.append(date)

    # Loop through all links then
    links = match.find_elements(By.TAG_NAME, 'a')  # Find all <a> tags
    for link in links:
        href = link.get_attribute('href')  # Get the hyperlink
        if href and href.endswith('match-summary'):  # Check if the link ends with 'match-summary'
            match_summary_links.append(href)  # Store the link in the list

            
# Quit the driver - closes web page
driver.quit()



import pandas as pd

# Create a DataFrame from data
data = {
    "date": date_ls,
    "home_team": home_team_ls,
    "away_team": away_team_ls,
    "home_score": home_score_ls,
    "away_score": away_score_ls,
    "url": match_summary_links
}

df = pd.DataFrame(data)

df.to_csv('flashscore_world_cup_2023.csv',index=False)


```

```python
# Now loop through match summary

# Set up Edge options
edge_options = Options()
edge_options.add_argument("--headless")  # Run in headless mode
edge_options.add_argument("--disable-gpu")
edge_options.add_argument("--no-sandbox")
edge_options.add_argument("--disable-dev-shm-usage")

# Set up the Edge WebDriver
webdriver_path = 'C:\\Users\\killi\\OneDrive\\Desktop\\edgedriver_win64\\msedgedriver.exe'
service = Service(webdriver_path)

# Create driver again - this opens up web page
driver = webdriver.Edge(service=service, options=edge_options)

url_list = df.url.tolist()
# Adjust url list
match_stats_url_list = [link + "/match-statistics/0" for link in url_list]

# Create dataframe to store all results
all_stats = pd.DataFrame()

j = 0
# Loop through each url and get the required data
for url in match_stats_url_list:
    driver.get(url)
    time.sleep(2)
    match_stats = driver.find_elements(By.CSS_SELECTOR, 'div.section')
    # Extract text from each element and combine it into a single string
    stats_string = "\n".join([stat.text for stat in match_stats])
    if "PRE-MATCH ODDS" in stats_string:
        stats_string = stats_string[:stats_string.find("PRE-MATCH ODDS")]

    lines = stats_string.splitlines()

    # Initialize variables
    attributes = []
    home_results = []
    away_results = []

    # Parse the lines into a structured format
    for i in range(0, len(lines), 3):  # Process in groups of 3 (home, metric, away)
        if i + 2 < len(lines):  # Ensure enough lines exist for home, metric, away
            home_results.append(lines[i])  # Home result
            attributes.append(lines[i+1])  # Metric/attribute
            away_results.append(lines[i+2])  # Away result

    # Create a DataFrame
    match_stats = {
        "url": url,
        "Attribute": attributes,
        "Home Result": home_results,
        "Away Result": away_results,
    }
    # Create dataframe for match stats
    match_stats_df = pd.DataFrame(match_stats)

    # Append to large dataframe
    all_stats = pd.concat([all_stats, match_stats_df], ignore_index=True)

    print(j,' loop complete')
    j = j + 1
    

# Save as csv
all_stats.to_csv('flashscore_world_cup_2023_all_stats.csv')


# Quit the driver - closes web page
driver.quit()

```
