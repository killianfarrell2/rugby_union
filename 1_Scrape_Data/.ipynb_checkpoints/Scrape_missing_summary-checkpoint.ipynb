```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.edge.service import Service
from selenium.webdriver.edge.options import Options
import time

# Read in Missing URLS

# Set folder
file_folder = 'C:/Users/killi/KF_Repo/PGA_Golf/Python_Scripts/Rugby_Union/2. Merge data/'
file = 'all_missing.csv'

missing_matches = pd.read_csv(file_folder+file)

# Create list from missing matches column
missing_matches_list = missing_matches['key'].to_list()

# Find how many matches are missing
print(len(missing_matches_list))
```

```python
# Now loop through Missing Matches

# Set up Edge options
edge_options = Options()
edge_options.add_argument("--headless")  # Run in headless mode
edge_options.add_argument("--disable-gpu")
edge_options.add_argument("--no-sandbox")
edge_options.add_argument("--disable-dev-shm-usage")

# Set up the Edge WebDriver
webdriver_path = 'C:\\Users\\killi\\OneDrive\\Desktop\\edgedriver_win64\\msedgedriver.exe'
service = Service(webdriver_path)

# Create driver again - this opens up web page
driver = webdriver.Edge(service=service, options=edge_options)

# Create dataframe to store all results
all_home_stats = pd.DataFrame()
all_away_stats =  pd.DataFrame()

# Loop through each url and get the required data

j = 0
for url in missing_matches_list:
    driver.get(url)
    time.sleep(2)

    # Get Home elements and away elements
    # Need to copy outer html, add div. and remove spaces
    home_elements = driver.find_elements(By.CSS_SELECTOR, 'div.smv__participantRow.smv__homeParticipant')
    away_elements = driver.find_elements(By.CSS_SELECTOR, 'div.smv__participantRow.smv__awayParticipant')

    # Create empty lists
    home_scores_list = []
    away_scores_list = []

    # Loop through home elements
    for home_element in home_elements:
        try:
            # Extract a sub-element, for example, a span or div within each home_element
            sub_element = home_element.find_element(By.CSS_SELECTOR, 'div.smv__subIncident')  # Change the selector as needed
    
            # You can now work with 'sub_element' - for example, get its text
            sub_element_text = sub_element.text
            # Remove brackets
            sub_element_text = sub_element_text.replace("(", "").replace(")", "")

            # Append the cleaned text to the list
            home_scores_list.append(sub_element_text)
    
        except Exception as e:
            # If the element is not found, skip this iteration
            continue  # Skip to the next iteration

    # Loop through away elements
    for away_element in away_elements:
        try:
            # Extract a sub-element, for example, a div within each away_element
            sub_element = away_element.find_element(By.CSS_SELECTOR, 'div.smv__subIncident')  # Change the selector as needed
        
            # Get the text of the sub-element
            sub_element_text = sub_element.text
        
            # Remove brackets from the text
            sub_element_text = sub_element_text.replace("(", "").replace(")", "")
        
            # Append the cleaned text to the list
            away_scores_list.append(sub_element_text)
    
        except Exception as e:
            # If the element is not found, skip this iteration
            continue  # Skip to the next iteration

    # Put home lists into dataframe
    home_list_df = pd.DataFrame(home_scores_list, columns=['Event'])
    home_event_counts = home_list_df['Event'].value_counts()
    home_event_counts =home_event_counts.to_frame().T
    home_event_counts['key'] = url

    # Put away list into df
    away_list_df = pd.DataFrame(away_scores_list, columns=['Event'])
    away_event_counts = away_list_df['Event'].value_counts()
    away_event_counts = away_event_counts.to_frame().T
    away_event_counts['key'] = url

    # Now append event counts to dataframes at the start with urls
    all_home_stats = pd.concat([all_home_stats, home_event_counts], ignore_index=True)
    all_away_stats = pd.concat([all_away_stats, away_event_counts], ignore_index=True)

    print(j,' loop complete')
    j = j + 1

# Quit the driver - closes web page
driver.quit()
    
```

```python
# Save down these dataframes and then join them back to missing data
# Fill NaN values with 0

all_away_stats= all_away_stats.fillna(0)
all_home_stats= all_home_stats.fillna(0)

# Save as csv
all_away_stats.to_csv('all_away_stats_missing.csv',index=False)
all_home_stats.to_csv('all_home_stats_missing.csv',index=False)
```
